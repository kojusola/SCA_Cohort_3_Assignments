#### DIABETES 
Diabetes is a disease in which the body doesn't produce enough or any insulin, doesn't properly use the insulin produced or exhibits a combination of both. When any of the above happens, the body is unable to get sugar from the blood into the cells. this leads to high blood sugar levels.
There are 3 major types of Diabetes:
1. Type 1 Diabetes
2. Type 2 Diabetes
3. Gestational Diabetes

Type 2 Diabetes is also called Adult onset diabetes, In that it majorly starts manifesting when the individual is an adult.

Risk factors for Type 2 diabetes:
a. Age(greater than 45 years old)
b. weight (greater 120% of the desirable body weight)
c. family history of Type 2 diabetes
d. History of impaired glucose tolerance or impaired fasting glucose
e. Hypertensio(>140/90)
f. History of gestational diabetes mellitus
g. Hispanic, African American, Asian American, Native American

### Creating new columns

```python
diabetes_data.loc[:,'Ng']=0
diabetes_data.loc[(diabetes_data['BloodPressure']<80),'Nbp']= 1
diabetes_data.loc[:,'Nbp']=0
diabetes_data.loc[(diabetes_data['Glucose']<105),'Ng']= 1
diabetes_data=diabetes_data.drop(['DiabetesPedigreeFunction'],axis=1)
```

```python
diabetes_data.head()
```

```python
diabetes_features = diabetes_data.drop(['Outcome'], axis=1)
diabetes_target = diabetes_data['Outcome']
```

```python
diabetes_features = pd.get_dummies(diabetes_features)
```

### Standardizing the data and model

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import OneHotEncoder,StandardScaler
from sklearn.model_selection import train_test_split
```

```python
train_data=StandardScaler().fit_transform(diabetes_features)
```

```python
X_train, X_test, y_train, y_test = train_test_split(train_data,diabetes_target,
                                                    test_size=0.30,random_state=1)
```

### Random forest classifier

```python
rf=RandomForestClassifier(random_state =1)
```

```python
rf.fit(X_train,y_train)
```

```python
Predictions = rf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy=accuracy_score(y_test,Predictions)
accuracy
```

### K Neighbors Classifier

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn =  KNeighborsClassifier(n_neighbors=5)
```

```python
knn.fit(X_train,y_train)
```

```python
y_pred = knn.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy=accuracy_score(y_test,y_pred)
accuracy
```
